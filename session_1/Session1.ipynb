```python
from Bio.Seq import Seq
from Bio.SeqUtils import nt_search
import matplotlib.pyplot as plt
from collections import Counter
```

```python
#1
#Seq A : AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC
#Seq B : CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG

Seq_A = Seq("AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC")
Seq_B = Seq("CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG")
print(len(Seq_A), len(Seq_B))
```

```python
#2
print(Seq_A.count("CGC"))
print(Seq_B.count("CGC"))
```

```python
#3
#Bisa pake find, bisa pake index
#Kalau pake find, dan gaketemu, -1 returned
#Dan find bisa dimulai dengan index custom
#Kalau pake index, dan gaketemu, keluaran error substring not found,
#Jadi harus pake if else (if "xxxx" in Seq_B:)
print(Seq_A.index("CAGTC"))
#print(Seq_A.find("CAGTC"))
print(Seq_B.find("CAGTC"))

SubSeq_A = "CAGTC"
SubSeq_B = "CAGTC"
print(nt_search(str(Seq_A), str(SubSeq_A)))
print(nt_search(str(Seq_B), str(SubSeq_B)))
```

```python
#4
#15 pertama dari seqA
#10 terakhir dari seqB
# Seq_AA = Seq_A[:15]

# len_SeqB = len(Seq_B)
# Seq_BB = Seq_B[len_SeqB-10:]

# Seq_C = Seq_AA + Seq_BB
#AGCTTGCAGCGTCCGTCGTTAGTGG

Seq_C = (Seq_A[:15] + Seq_B[len(Seq_B)-10:])
print(Seq_C)
print(Seq_C[1:6])
```

```python
#5
Seq_C_Reverse = Seq_C[::-1]
print(Seq_C_Reverse)
```

```python
#6
freq_Seq_A = Counter(Seq_A)
freq_Seq_B = Counter(Seq_B)
freq_Seq_C = Counter(Seq_C)

freqKeys_Seq_A = freq_Seq_A.keys()
freqKeys_Seq_B = freq_Seq_B.keys()
freqKeys_Seq_C = freq_Seq_C.keys()

freqValues_Seq_A = freq_Seq_A.values()
freqValues_Seq_B = freq_Seq_B.values()
freqValues_Seq_C = freq_Seq_C.values()

plt.bar(freqKeys_Seq_A, freqValues_Seq_A)
plt.title("Seq_A")
plt.show()

plt.bar(freqKeys_Seq_B, freqValues_Seq_B)
plt.title("Seq_B")
plt.show()

plt.bar(freqKeys_Seq_C, freqValues_Seq_C)
plt.title("Seq_C")
plt.show()
```
