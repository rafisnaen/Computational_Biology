```python
from Bio.Seq import Seq
```

```python
Seq_A = Seq("AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC")
Seq_B = Seq("CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG")
```

```python
from Bio.SeqUtils import molecular_weight

Weight_Seq_A = molecular_weight(Seq_A)
Weight_Seq_B = molecular_weight(Seq_B)
print(Weight_Seq_A, "", Weight_Seq_B)
```

```python
from Bio.SeqUtils import MeltingTemp
import matplotlib.pyplot as plt

MT_Seq_A = MeltingTemp.Tm_Wallace(Seq_A)
MT_Seq_B = MeltingTemp.Tm_Wallace(Seq_B)
print(MT_Seq_A, " ", MT_Seq_B)

plt.bar(["Seq A MT", "Seq B MT"], [MT_Seq_A, MT_Seq_B])
plt.xlabel("Sequences")
plt.ylabel("Melting Temperature")
plt.title("Sequence A & B Melting Temperature")
plt.show()
```

```python
from Bio.SeqUtils import GC
Seq_A_GC_Comp = GC(Seq_A)
Seq_A_AT_Comp = 100 - Seq_A_GC_Comp

Seq_B_GC_Comp = GC(Seq_B)
Seq_B_AT_Comp = 100 - Seq_B_GC_Comp

print(Seq_A_GC_Comp, Seq_A_AT_Comp)
print(Seq_B_GC_Comp, Seq_B_AT_Comp)
```

```python
plt.bar(["Seq A GC Comp", "Seq A AT Comp", "Seq B GC Comp", "Seq B AT Comp"], [Seq_A_GC_Comp, Seq_A_AT_Comp, Seq_B_GC_Comp, Seq_B_AT_Comp])
plt.xlabel("Sequences")
plt.ylabel("GC AT Composition")
plt.title("GC AT Composition of Sequence A and B")
plt.show()
```
